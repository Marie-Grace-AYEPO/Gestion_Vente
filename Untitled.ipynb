```python
from datetime import datetime
import pandas as pd

# Convertir la colonne 'dateachat' en datetime
df['dateachat'] = pd.to_datetime(df['dateachat'])

# Insérer les données dans la table 'client', 'categorie', 'produit' et 'vente'
for i in df.index:
    # Insérer les données dans la table 'client'
    cursor.execute("""
    INSERT INTO client (nom, prenoms, genre, telephone, adresse) 
    VALUES (?, ?, ?, ?, ?)
    """, 
    df['nom'][i], df['prenoms'][i], df['Genre'][i], np.random.randint(10000000,99999999), 'Abidjan') 

    # Récupérer l'id du client inséré
    id_client = cursor.execute("SELECT @@IDENTITY AS 'Identity' ").fetchval()

    # Insérer les données dans la table 'categorie'
    cursor.execute("""
    INSERT INTO categorie (nom) 
    VALUES (?)
    """, df['categorie'][i]) 

    # Récupérer l'id de la catégorie insérée
    id_categorie = cursor.execute("SELECT @@IDENTITY AS 'Identity' ").fetchval()

    # Insérer les données dans la table 'produit'
    cursor.execute("""
    INSERT INTO produit (nom, descriptions, prixunitaire, id_categorie) 
    VALUES (?, ?, ?, ?)
    """, 
    df['nomproduit'][i], 'description', int(df['prixunitaire'][i]), id_categorie) 

    # Récupérer l'id du produit inséré
    id_produit = cursor.execute("SELECT @@IDENTITY AS 'Identity' ").fetchval()

    # Insérer les données dans la table 'vente'
    cursor.execute("""
    INSERT INTO vente (id_vendeur, id_client, id_produit, quantite, montant, dateachat) 
    VALUES (?, ?, ?, ?, ?, ?)
    """, 
    1, id_client, id_produit, 1, int(df['prixunitaire'][i]), df['dateachat'][i]) 

conn.commit()
```
